```python
import pandas as pd 
from pathlib import Path
```

```python
folder_path = Path(r"C:\Users\F0040RP\Documents\DartLib_RDS\persConsultations\WoS\resilience")
```

```python
df = pd.read_csv(Path(folder_path, "resilience1_500.txt"), encoding = "utf-8", sep = "\t")
df.head()
```

```python
df.loc[0:5, 'AU']
```

```python
subdf = df.loc[:, ['AU', 'TI', 'SO', 'UT', 'CR', 'NR', 'TC', 'Z9']]
subdf.shape
```

```python
subdf.head()
```

```python
authors_only_df = subdf['AU'].str.split("; ", expand = True)
```

```python
authors_only_df.describe()
```

```python
authors_only_df.columns = [str(int(col) + 1) for col in authors_only_df.columns]
authors_only_df.head(2)
```

```python
authorsdf = pd.concat([subdf, authors_only_df], axis = 1)
authorsdf.head(2)
```

```python
authors_longdf = pd.melt(authorsdf, id_vars=['AU', 'TI', 'SO', 'UT', 'CR', 'NR', 'TC', 'Z9'])
```

```python
authors_longdf.tail()
```

```python
authors_longdf = authors_longdf.rename(columns = {"variable": "author_pos", "value": "author"})
authors_longdf.head(3)
```

```python
print(authors_longdf.shape)
authors_longdf = authors_longdf.dropna(subset = ["author"])
print(authors_longdf.shape)
```

```python
authors_longdf['num_authors'] = authors_longdf['AU'].apply(lambda x: len(x.split("; ")))
authors_longdf.head()
```

```python
authors_longdf = authors_longdf.sort_values(by = ['UT', 'author_pos'])
authors_longdf.head(10)
```

```python
authors_longdf.to_csv(Path(folder_path, "resilience_author-pos.csv", index = False), encoding = 'utf-8')
```
