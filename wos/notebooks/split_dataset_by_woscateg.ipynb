## Explode and export full dataset

```python
csv_dir = Path("../data/merged_wos_files.csv")                        # now apply to full dataset
df = pd.read_csv(csv_dir, encoding = "utf-8", index_col = [0])

# split string of WoS Categories in column "WC" into a list  
df['wos_categs'] = df['WC'].str.split(';')                                                                       

# explode WoS categ's for each paper into multiple rows; i.e. a row/paper w 3 WC categories is split into 3 rows with 1 WC each
df_wc_explode = df.explode("wos_categs")  
#strip leading / trailing whitespace                         
df_wc_explode['wos_categs'] = df_wc_explode['wos_categs'].str.strip()


```

```python
df_wc_explode['wos_categs'] = df_wc_explode['wos_categs'].fillna("")
```

```python
# strip leading / trailing punctuation
#df_wc_explode['wos_categs'] = df_wc_explode['wos_categs'].str.translate(str.maketrans('', '', string.punctuation))   #this strips all punctuation, don't want that!
def strip_outer_punct(txt: str):
    newtxt = re.sub(r"^[\W\s]*(.*?)[\W\s]*$", r"\1", txt)
    if newtxt != txt:
        print(newtxt)
    return(newtxt)
df_wc_explode['wos_categs'] = df_wc_explode['wos_categs'].apply(strip_outer_punct)

```
