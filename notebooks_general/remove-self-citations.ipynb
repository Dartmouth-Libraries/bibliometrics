```python
#def count_self_citations(CRlist: list, authorname: str):
def count_self_citations(CRs: str, authorname: str):
    """
    This function reads in:
    1. a list of cited references (from WoS column "CR"), with each reference separated by a ";"
    2. an author's name, which we split out from the column "AU" into the column "author"
    """
    #print(CRs)
    CRlist = CRs.split("; ")
    #print(authorname)
    authorname_split = authorname.split(", ")
    #print(authorname_split)
    if len(authorname_split) > 1:
        #authorname2 = " ".join(authorname_split)
        authorname2 = authorname_split[0] + " " + authorname_split[1][0]
        #print("an2:", authorname2)
        if len(authorname_split) > 2:
            print("LONG NAME: ", authorname_split)
    else:
        authorname2 = authorname
    authorname2_length = len(authorname2)
    CR_authorlist = []
    ctr = 0
    for i, CR in enumerate(CRlist):
        #print(CR)
        CR_author = CR.split(", ")[0]
        #print(CR_author, "/", authorname2)
        CR_authorlist.append((i, CR))
        #if authorname2 == CR_author[ : authorname2_length]:
        if CR_author.startswith(authorname2):
            print("Match: ", authorname, "in:", CRlist[i])
            ctr += 1
    return(ctr)
        
    
```
